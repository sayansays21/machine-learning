# extracting data and building initial table

```python
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples=100,n_features=2,centers=2,random_state=19)
import pandas as pd
import numpy as np

#mentioning the args
eps=5
min_pt=2

d = {}
d['Points']=list(x)
d['Cluster']=0
d['Reachability Distance']=1000
d['Core Distance']=1000
df = pd.DataFrame(d)
df
```

# function to find neighbourhood

```python

def neigh_point(p):
    n_points=[]
    for i in range(len(x)):
        if i!=p:
            a=np.linalg.norm(x[p]-x[i])
            if a<eps:
                n_points.append([x[p],i,a])
    return(pd.DataFrame(n_points,columns=["point1","neighbourhoods","distance"]))

s=neigh_point(0)
s.sort_values("distance")
```

# updating cluster, core distance and reachable distance

```python
def update():
    cluster=0
    for i in range(len(x)):
        v=neigh_point(i)
        v=v.sort_values("distance")
        if (df.iloc[i,1]==0 and len(v)>=min_pt):
            df.iloc[i,3]=v.iloc[min_pt-1,2]
            cluster=cluster+1
            df.iloc[i,1]=cluster
            for j in range(len(v)):
                rea_dis=max(df.iloc[i,3],v.iloc[j,2])
                if (rea_dis<df.iloc[v.iloc[j,1],2]):
                    df.iloc[v.iloc[j,1],2]=rea_dis
                w=neigh_point(j)
                w=w.sort_values("distance")
                if (df.iloc[v.iloc[j,1],1]==0 and len(w)>=min_pt):
                    df.iloc[v.iloc[j,1],3]=w.iloc[min_pt-1,2]
                    df.iloc[v.iloc[j,1],1]=cluster
                for h in range(len(w)):
                    rea_dis=max(df.iloc[i,3],w.iloc[h,2])
                    if (rea_dis<df.iloc[w.iloc[h,1],2]):
                        df.iloc[w.iloc[h,1],2]=rea_dis
                
```

```python
update() 
```
